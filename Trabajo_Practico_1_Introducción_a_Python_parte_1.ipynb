### **Ejercicio 29**

Construir un programa que lea un número natural **N** y calcule la suma de los primeros **N** números pares.

```python
nroNatural = int(input("Ingrese un numero natural: "))

suma = 0

for i in range(2, 2*nroNatural+1, 2):
    if i % 2 == 0:
        suma += i

print(suma)
```

```python
nroUno = int(input("Ingrese un numero entero: "))
nroDos = int(input("Ingrese otro numero entero: "))


if nroUno < nroDos:
    aux = nroUno+1
    while aux < nroDos:
        if aux % 2 == 0:
            print(aux)

        aux += 1

            
```
